```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 1.1 Load data set
df = pd.read_csv('Heart Disease UCI.csv')
df.info()
```

```python
# 1.2 Displaying first 10 data set 

df.head(10)
```

```python
# 1.3 Checking Data types

df.dtypes
```

```python
# 1.4 DATA CLEANING

df.isnull().sum()
df.fillna(df.mean(), inplace=True)
df.isnull().sum()
```

```python
# 1.5 Display Summary Statistics

df.describe()
```

```python
# 2.1 Count number of Males and number of Females

df['sex'].value_counts()
```

```python
#  2.2 Calculating Percentage Distribution

percentage = df['condition'].value_counts()

c =percentage/len(df)*100
print("Percentage:")
print(percentage, c.round(2))
plt.bar(percentage.index.astype(str),c)
plt.xlabel("Condition (Target)")
plt.ylabel("Percentage")
plt.title("Percentage Distribution based on condition")
plt.show()
```

```python
#  3.1 min, max, mean, median of age

mn = df['age'].min()
mx = df['age'].max()
men = df['age'].mean()
med = df['age'].median()
print("Min:",mn)
print("Max:",mx)
print("Mean:",men.round(2))
print("Median:",med)    
```

```python
# 3.2 age_distribution = df['age'].value_counts()

plt.figure(figsize=(8,5))
plt.hist(df['age'], bins=10, color='lightgreen', edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
# 4.1 Count patients with and without heart disea
heart_disease = df['condition'].value_counts()
print("Heart disease counts (0=no, 1=yes):")
print(heart_disease)

#4.2  pie chart
plt.figure(figsize=(5,5))
labels = ['No Disease', 'Disease'] if len(heart_disease)==2 else heart_disease.index.astype(str)
plt.pie(heart_disease, labels=labels, autopct='%1.1f%%', colors=['lightgreen','salmon'])
plt.title('Patients with and without Heart Disease')
plt.show()
```

```python
# 4.3 Calculate disease percentage 
percent =  df['condition'].value_counts()
d = percent/len(df) * 100
print("Disease Percentage:",d)

plt.bar(percent.index.astype(str),d)
plt.xlabel("Disease Target")
plt.ylabel("Percentage of disease")
plt.title("Disease percentage")
plt.show()
print(d)
```

```python
# 5.1 Correlation between age and cholesterol #scatter
corr = df.corr(numeric_only=True)
print(corr)
```

```python
# 5.2 Age vs cholestrol Scatter plot

plt.figure(figsize=(8,5))
plt.scatter(df['age'], df['chol'], alpha=0.6, edgecolor='k')
plt.title('Scatter Plot: Age vs Cholesterol')
plt.xlabel('Age')
plt.ylabel('Cholesterol')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

```python
# 5.3 Interpret relationship 

corr_value = corr.loc['age','chol'] if 'age' in corr.index and 'chol' in corr.columns else None
if corr_value is not None:
    print(f"Pearson correlation between age and cholesterol: {corr_value:.3f}")
    if corr_value > 0:
        print("There is a positive relationship: as age increases, cholesterol generally rises.")
    elif corr_value < 0:
        print("There is a negative relationship: older patients tend to have lower cholesterol.")
    else:
        print("No linear relationship detected.")
else:
    print("Correlation value for age vs chol was not available.")

```

```python
# INTERMEDIATE 

# 6.1 Chest pain type vs disease rate
grouped = df.groupby('cp')['condition'].mean() * 100  
print("Disease rate by chest pain type (percentage):")
print(grouped)

# 6.2 bar chart of rates
plt.figure(figsize=(8,5))
grouped.plot(kind='bar', color='orange', edgecolor='black')
plt.title('Heart Disease Rate by Chest Pain Type')
plt.xlabel('Chest Pain Type (cp)')
plt.ylabel('Disease Rate (%)')
plt.show()

# 6.3 Identify which chest pain type is most risky 
max_disease_rate = grouped.max()
print("Most risky chest pain type (highest disease rate):", grouped.idxmax(), "with a disease rate of", max_disease_rate.round(2), "%")
```

```python
# 7.1 Average Cholesterol by Gender
# 7.2 Mean Cholesterol 
grouped = df.groupby('sex')['chol'].mean()*100 #value printing

# 7.3 Bar chart of average cholesterol
plt.figure(figsize=(8,5))
grouped.plot(kind='bar', color='pink', edgecolor='white')
plt.title('Average Cholestrol by gender')
plt.xlabel('Gender')
plt.ylabel('Cholestrol')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```

```python
# 8.1 Resting blood pressure analysis
# Average BP

avg_bp = df['trestbps'].mean()

print(f"Average resting blood pressure: {avg_bp:.2f}")
```

```python
# 8.2 High BP

high_bp = df[df['trestbps'] > 140]
count_high_bp = len(high_bp)

print(f"Number of patients with BP > 140: {count_high_bp}")
```

```python
# 8.3 compare disease presence in high BP group
disease_rate_high = high_bp['condition'].value_counts(normalize=True) * 100
print("\nDisease distribution among high BP patients (%):")
print(disease_rate_high)

plt.figure(figsize=(6,4))
disease_rate_high.plot(kind='bar', color=['red','lightgreen'], edgecolor='black')
plt.title('Disease Presence in High BP (>140) Group')
plt.xlabel('Heart Disease (0=no,1=yes)')
plt.ylabel('Percentage')
plt.show()
```

```python
# 9.1 Maximum heart rate (thalach) comparison by disease status
disease_group = df[df['condition'] == 1]['thalach'].mean()
non_disease_group = df[df['condition'] == 0]['thalach'].mean()

print(f"Average thalach for disease patients: {disease_group:.2f}")
print(f"Average thalach for non-disease patients: {non_disease_group:.2f}")
```

```python
# 9.2 Boxplot

plt.figure(figsize = (4,4))
df.head(5)
sns.boxplot(hue='condition', y='thalach', data=df, palette=['lightcoral','lightblue'])
plt.title('Max Heart Rate by Disease Status')
plt.xlabel('Heart Disease (0=no,1=yes)')
plt.ylabel('Thalach (max heart rate)')
plt.show()
```

```python
# 10.1 Exercise Induced Angina Impact 
# calculate percent with disease in each exang group

rates = df.groupby('exang')['condition'].mean() * 100
print("Disease percentage by exang value:")
print(rates)

# 10.2 11bar chart 
plt.figure(figsize=(6,4))
rates.plot(kind='bar', color=['green','red'], edgecolor='black')
plt.title('Exercise Induced Angina Impact on Heart Disease')
plt.xlabel('Exang')
plt.ylabel('Percentage with Disease')
plt.show()
```

```python
# 11.1 Number of Major Vessels (ca) Impact 
# ST depression (oldpeak) analysis
mean_oldpeak = df.groupby('condition')['oldpeak'].mean()
print(mean_oldpeak)

# 11.2 histogram
plt.figure(figsize=(8,5))
for label, grp in df.groupby('condition'):
    plt.subplot(2,2,label+1)
    plt.hist(grp['oldpeak'], bins=20, alpha=0.5, label=f'target={label}')
plt.title('Oldpeak Distribution by Target')
plt.xlabel('Oldpeak')
plt.ylabel('Count')
plt.legend()
plt.show()
```

```python
# 11.3 Identify Trend
# compare the mean oldpeak values between groups
mean_vals = df.groupby('condition')['oldpeak'].mean()
print("Mean oldpeak by condition:")
print(mean_vals)

if mean_vals.diff().iloc[-1] > 0:
    print("Patients with heart disease tend to have higher oldpeak values.")
else:
    print("Patients without heart disease have similar or higher oldpeak values.")
```

```python
# 12.1 Number of major vessels (ca) impact
ca_rates = df.groupby('ca')['condition'].mean() * 100  # disease probability
print(ca_rates)

# line chart
plt.figure(figsize=(8,5))
ca_rates.plot(kind='line', marker='o', linestyle='-')
plt.title('Disease Probability vs Number of Major Vessels')
plt.xlabel('Number of major vessels (ca)')
plt.ylabel('Disease Probability (%)')
plt.show()
```

```python
# 13.1 Thalassemia vs disease
ct = pd.crosstab(df['thal'], df['condition'], normalize='index') * 100
print(ct)

# stacked bar chart
plt.figure(figsize=(8,5))
ct.plot(kind='bar', stacked=True, color=['lightgreen','salmon'], edgecolor='black')
plt.title('Thalassemia Type vs Disease Percentage')
plt.xlabel('Thalassemia Type')
plt.ylabel('Percentage')
plt.legend(title='Disease')
plt.tight_layout()
plt.show()
```

```python
# 14.1 Multi-factor risk analysis using NumPy filtering

criteria = (df['age'] > 50) & (df['chol'] > 240) & (df['trestbps'] > 140)
risky = df[np.array(criteria)]

total_risky = len(risky)
disease_risky = risky['condition'].sum()  
percent_disease = disease_risky / total_risky * 100 if total_risky > 0 else 0

print(f"Total patients meeting criteria: {total_risky}")
print(f"Number with disease among them: {disease_risky}")
print(f"Percentage with disease: {percent_disease:.2f}%")
```

```python
# 15.1 Custom risk score creation   double check

df['risk_score'] = (df['chol']/200) + (df['trestbps']/120) + df['oldpeak']

# 15.2 classify into risk categories
bins = [0, 2, 4, np.inf]
labels = ['Low Risk', 'Medium Risk', 'High Risk']
df['risk_category'] = pd.cut(df['risk_score'], bins=bins, labels=labels)

print(df[['risk_score', 'risk_category']].head())

# 15.3 visualize distribution of categories
plt.figure(figsize=(6,4))
sns.countplot(x='risk_category',hue='risk_category', legend = False,data=df, palette='viridis')
plt.title('Risk Category Distribution')
plt.xlabel('Risk Category')
plt.ylabel('Count')
plt.show()
```
