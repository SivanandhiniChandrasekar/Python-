```python
# iterator
It is an object that allows you to traverse through all the elements of a collections (like lists or tuples) one a a time.
__iter__() and __next__() methods.

Raises stop iteration when no more data is available.
Used in loops like for loop.
  
```

```python
import time

start = time.perf_counter()

class mynum:
    def __iter__(self):
        self.num = 1
        return self
    def __next__(self):
        if self.num<=5:
            x = self.num
            self.num += 1
            return x
        else:
            raise StopIteration
ob = mynum()
myiter = iter(ob)
for x in myiter:
    print(x)
end = time.perf_counter()

print(f"Execution time: {end - start:.6f} seconds")
```

```python
A generator is a special type of iterator that is defined using a function and the yield keyword.

-> uses yield instead of return 
-> Generator values lazily (ones at a time)
-> more memory efficient than normal functions.
```

```python
import time

start = time.perf_counter()

def my_generator():
    for i in range(1,6):
        yield i
gen = my_generator()
for value in gen:
    print(value)


end = time.perf_counter()

print(f"Execution time: {end - start:.6f} seconds")
```

```python
Generator vs Iterator
Iterator -> Requires class with __iter__() and __next__()
         -> More code, manual implementations
         -> More memory efficient

Generator -> Uses function with yield keyword
          -> Simple and concise implementation
          -> Slightly more control  
```
