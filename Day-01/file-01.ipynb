```python
# prime number
a = int(input())
flag = 0
if a == 1 or a == 2:
    print("Prime")
elif a > 2:
    for i in range(2, a):
        if a % i == 0:
            flag = 1
            print("Not a prime")
            break # it doesn't execute the records after the break statement.
    if flag == 0:
        print("Prime")
   
    
```
