```python
def hcf(a,b):
    while a>0:
        a,b = b%a,a
    return b if b >= 1 else 1

a = int(input())
b = int(input())
res = hcf(max(a,b),min(a,b))
print(res)
```

```python
# MAP
# Applies a function to all the items in a list

a = [1,2,3,4,5]
print(list(map(lambda x:x**2, a)))
```

```python
#  REDUCE
#  it is for performing some computation a list and returning a list
# it applies a rolling computation to sequential pairs of values 
from functools import reduce
a = [1,2,3,4,5,6]
b = reduce(lambda x,y:x*y, a)
print(b)
```

```python
"""
Keyword Argument - kwargs is used if you want to handle named arguments in function
"""
def lst(**kwargs):
    if kwargs:
        return (f"Name: {kwargs['name']}, Mob_No:{kwargs['mob']},DOB: {kwargs['dob']}")
    else:
        return "No arguments"
print(lst(name = "Nand", mob = "8973920239", dob = "23/09/2001"))
```

```python
"""
Arbitrary Argument - it all to pass n no.of arguments pass into a function
"""

def lst(*args):
    if args:
        print( args)
        for i in args:
            print( i)
lst("Siva",20,90,34,54,90)
```

```python
class member:
    def __init__(self, name, age, position):
        self.name = name
        self.age = age
        self.position = position
    def hi(self, name, age):
        self.name = name
        self.age = age
        
        

test = member("Joe",27,"Developer")
print(test.name)
print(test.age)
test2 = member("Mad",27, "Servicenow")
print(test2.name)
print(test2.age)
print(test.hi)
```

```python
class member:
    def __init__(self, name, age, position):
        self.name = name
        self.age = age
        self.position = position
    def testfun(self):
        return ("Hello my name is "+self.name)
        

test = member("Joe",27,"Developer")
print(test.name)
print(test.age)
print(test.position)
print(test.testfun())
```
