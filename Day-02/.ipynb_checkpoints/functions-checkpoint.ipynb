```python
def get_sum(lst):
    """
    The function adds array elements and return their sum
    """
    all = 0
    for num in lst:
        all += num
        print(all)
    return all
lst2 = get_sum([1,2,3,4])
print(lst2)
print(get_sum,__doc__)
```

```python
def hcf(a,b):
    while a>0:
        a,b = b%a,a
    return b if b >= 1 else 1

a = int(input())
b = int(input())
res = hcf(max(a,b),min(a,b))
print(res)
```

```python
# Using lambda function to get positive numbers in an array

lst1 = [-1,-2,-3,9,8,7,5,4,3]
lst2 = list(filter(lambda n:n>0, lst1))
print(lst2)
```

```python
print(list(filter(lambda n:n<0,lst1)))
```

```python
# MAP
# Applies a function to all the items in a list

a = [1,2,3,4,5]
print(list(map(lambda x:x**2, a)))
```

```python
#  REDUCE
#  it is for performing some computation a list and returning a list
# it applies a rolling computation to sequential pairs of values 
from functools import reduce
a = [1,2,3,4,5,6]
b = reduce(lambda x,y:x*y, a)
print(b)
```
