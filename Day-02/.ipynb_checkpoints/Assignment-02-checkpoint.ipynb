```python
# 1. We can't call local variable from outside because of the scope rules.
# The local variable is accessed only when the function is called.
```

```python
Map - if we want all the elements in an array to perform same operation.
    Example: [1,2,3] => Square of each element in an array 
    d = list(map(lambda x:x**2 ,[1,2,3]))
    print(d)
Filter - if we want to filter elements in an array that satisfy a condition.
    Example: [1,2,3] => if we want odd elements in an array
    d = list(filter(lambda x:x%2 ==1,[1,2,3]))
    print(d)
reduce - if we want single value i.e combine all elements into one result.
    Example: [1,2,3] => if we want sum of the array.
    from functools import reduce
    d = reduce(lambda x,y:x+y,[1,2,3])
    print(d)
```

```python
from functools import reduce
d = reduce(lambda x,y:x+y,[1,2,3])
print(d)
```
