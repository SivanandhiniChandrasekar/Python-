```python
import pandas as pd
import sqlite3

# Extract
data = pd.read_csv(r"Downloads/sales_data.csv")
print("Extracted Data:")
print(data)
```

```python
print(data.dtypes)
```

```python
# Remove null values
data = data.dropna()

data["Sale_Date"] = pd.to_datetime(data["Sale_Date"])
```

```python
data["Discount"] = data["Discount"].fillna(0)
```

```python
data["Total_Sales"] = data["Quantity_Sold"] * data["Unit_Price"]
```

```python
data["Profit"] = (data["Unit_Price"] - data["Unit_Cost"]) * data["Quantity_Sold"]
```

```python
data["Final_Amount"] = data["Total_Sales"] - data["Discount"]
```

```python
data[["Region_New","Sales_Rep_New"]] = data["Region_and_Sales_Rep"].str.split("-", expand=True)
```

```python
import sqlite3

conn = sqlite3.connect("sales.db")

data.to_sql("sales_table", conn, if_exists="replace", index=False)

conn.close()
```
