```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

months = ['jan','feb','mar','apr','may','jun']
traffic = [150,2,180,220,250,310]

plt.plot(months, traffic)
plt.show()
```

```python
plt.xlabel('Month')
plt.ylabel('Monthly Traffic')
plt.title('Monthly Website Traffic')
```

```python
plt.plot(months, traffic, marker='<',markersize = 10,color='pink', linestyle='-',linewidth=1)
```

```python
months = ['Jan','Feb','Mar','Apr','May','Jun']
product_a = [45,55,40,70,80,75]
product_b = [35,40,50,55,70,68]

plt.plot(months, product_a, marker = 's', linestyle = '-', color = 'blue', label = 'Product A')
plt.plot(months, product_b, marker = 'p', linestyle = '--', color = 'red', label = 'Product B')

plt.xlabel("Month")
plt.ylabel("Monthly Revenue(in $1000)")
plt.title("Monthly Revenue Comparison")

plt.legend()

plt.show()
```

```python
year = [2004, 2005, 2006, 2007, 2008, 2009]
hike = [10, 21, 25, 27, 32, 40]
plt.plot(year, hike)
plt.show()
```

```python
timi = [300000, 330000, 350000, 380000, 420000, 500000]
john = [420000, 450000, 470000, 480000, 510000, 530000]
abraham = [410000, 440000, 440000,500000, 520000,540000]
plt.plot(year, timi, marker = "*", linestyle ="-", color = 'blue', label = "Timi's Salary Hike")
plt.plot(year, john, marker = "o", linestyle ="-", color = 'orange', label = "John's Salary Hike")
plt.plot(year, abraham, marker = "s", linestyle ="-", color = 'pink', label = "Abraham's Salary Hike")
plt.legend()
plt.show()
```

```python
categories = ['category A','category B','category C','category D']
values = [20,23,30,35]

# bar - default vertical adding 'h' to bar to make it in horizontal view
plt.barh(categories, values)
         # , color = ['red','g','blue','pink'], width = 0.8)

plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar plot Example')

plt.show()
```
