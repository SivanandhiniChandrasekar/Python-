```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

months = ['jan','feb','mar','apr','may','jun']
traffic = [150,2,180,220,250,310]

plt.plot(months, traffic)
plt.show()
```

```python
plt.xlabel('Month')
plt.ylabel('Monthly Traffic')
plt.title('Monthly Website Traffic')
```

```python
plt.plot(months, traffic, marker='<',markersize = 10,color='pink', linestyle='-',linewidth=1)
```

```python
months = ['Jan','Feb','Mar','Apr','May','Jun']
product_a = [45,55,40,70,80,75]
product_b = [35,40,50,55,70,68]

plt.plot(months, product_a, marker = 's', linestyle = '-', color = 'blue', label = 'Product A')
plt.plot(months, product_b, marker = 'p', linestyle = '--', color = 'red', label = 'Product B')

plt.xlabel("Month")
plt.ylabel("Monthly Revenue(in $1000)")
plt.title("Monthly Revenue Comparison")

plt.legend()

plt.show()
```

```python
year = [2004, 2005, 2006, 2007, 2008, 2009]
hike = [10, 21, 25, 27, 32, 40]
plt.plot(year, hike)
plt.show()
```

```python
wins = df['winner'].value_counts()

plt.figure(figsize=(14,6))
plt.bar(wins.index, wins.values, color='skyblue')
plt.xlabel("Team")
plt.ylabel("Number of Wins")
plt.title("Most Matches Won by Team")
plt.xticks(rotation=90)
plt.show()
```

```python
# HISTOGRAM -> same as bar chart
import matplotlib.pyplot as plt
import numpy as np

data = np.random.random_sample(5)
plt.hist(data, color = "cyan", edgecolor = "white", bins = 5)

plt.xlabel("Values")
plt.ylabel("Frequency")
plt.title("Histogram plot")

plt.show()
```
