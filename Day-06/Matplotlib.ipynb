```python
# HISTOGRAM -> same as bar chart
import matplotlib.pyplot as plt
import numpy as np

data = np.random.random_sample(5)
plt.hist(data, color = "cyan", edgecolor = "white", bins = 5)

plt.xlabel("Values")
plt.ylabel("Frequency")
plt.title("Histogram plot")

plt.show()
```

```python
# SCATTER PLOT

x = [14,25,33,90]
y = [56,78,34,23]

plt.scatter(x,y, marker = "o", color = "red")

plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("Scatter Plot")

plt.show()
```

```python
# PIE CHART

categories = ['Housing','Transportation','Food','Entertainment','Utilities']
expenses = [1200,300,400,200,150]

plt.figure(figsize = (5,5))
plt.pie(expenses, labels = categories, autopct = "%2.2f%%")

plt.title("Monthly expenses by category")
plt.legend()
plt.show()
```

```python
data = [12,13,14,5,6,67,78,67,56,4,3,34,556]

plt.boxplot(data)

plt.title("Boxplot")
plt.xlabel("Data")
plt.ylabel("Values")

plt.show()
```

```python
quarters = ['Q1','Q2','Q3','Q4']
electronics = [10000, 12000, 11000,10500]
clothing = [5000, 6900, 7500, 8000]
homeAppliances = [3000, 7500, 8200, 9000]

plt.figure(figsize=(10,6))
plt.stackplot(quarters, electronics, clothing, homeAppliances, labels = ['Electronics','Clothing','Home Appliances'])

plt.xlabel("Quarters")
plt.ylabel("Sales ($)")
plt.title("Product Category Sales over Quarters")
plt.legend()
# plt.grid()
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-5,5,150)
y = np.linspace(-5,5,150)

x,y = np.meshgrid(x,y)
z = np.sin(np.sqrt(x**2 + y**2))

fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')
ax.scatter(x,y,z, c = z, cmap = 'viridis')
ax.set_xlabel('X-Axis')
ax.set_ylabel('Y-Axis')
ax.set_zlabel('Z-Axis')

plt.show()
```

```python
x = np.linspace(0,10,100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = np.exp(-x)

fig,axs = plt.subplots(2,2, figsize=(10,8))
axs[0,0].plot(x,y1,'r')
axs[0,0].set_title('Sine Wave')
axs[0,1].plot(x,y2,'b')
axs[0,1].set_title('Cosine Wave')
axs[1,0].plot(x,y3,'g')
axs[1,0].set_title('Tangent Wave')
axs[1,1].plot(x,y4,'m')
axs[1,1].set_title('Exponential Wave')

plt.tight_layout()
plt.show()
```
